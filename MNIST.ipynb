```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, regularizers
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import time
import os
import joblib

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

def load_and_prepare_data():
    """Load and prepare MNIST dataset for neural network training."""
    print("Loading MNIST dataset...")
    # Load data from OpenML
    X, y = fetch_openml('mnist_784', version=1, return_X_y=True, as_frame=False)
    
    # Convert to numpy arrays and ensure correct types
    X = np.array(X, dtype=np.float32)
    y = np.array(y, dtype=np.uint8)
    
    # Reshape data for CNN input (samples, height, width, channels)
    X = X.reshape(-1, 28, 28, 1)
    
    # Normalize pixel values to [0, 1]
    X = X / 255.0
    
    # Convert labels to one-hot encoding
    y = keras.utils.to_categorical(y, 10)
    
    # Split the data into training, validation, and test sets
    X_temp, X_test, y_temp, y_test = train_test_split(
        X, y, test_size=0.1, random_state=42, stratify=np.argmax(y, axis=1)
    )
    
    X_train, X_val, y_train, y_val = train_test_split(
        X_temp, y_temp, test_size=0.1, random_state=42, stratify=np.argmax(y_temp, axis=1)
    )
    
    print(f"Data loaded and prepared:")
    print(f"  - Training: {X_train.shape[0]} samples")
    print(f"  - Validation: {X_val.shape[0]} samples")
    print(f"  - Test: {X_test.shape[0]} samples")
    
    return X_train, X_val, X_test, y_train, y_val, y_test

def create_data_generators(X_train):
    """Create data generators for data augmentation."""
    # Data augmentation for training
    train_datagen = ImageDataGenerator(
        rotation_range=10,  # randomly rotate images by up to 10 degrees
        width_shift_range=0.1,  # randomly shift images horizontally by up to 10% of width
        height_shift_range=0.1,  # randomly shift images vertically by up to 10% of height
        zoom_range=0.1,  # randomly zoom images by up to 10%
        shear_range=0.1,  # shear intensity
        fill_mode='nearest'  # strategy for filling points outside the input boundaries
    )
    
    # Setup generator for training data
    train_datagen.fit(X_train)
    
    return train_datagen

def build_cnn_model():
    """Build a convolutional neural network model for MNIST."""
    model = keras.Sequential([
        # First convolutional block
        layers.Conv2D(32, kernel_size=(3, 3), activation='relu', padding='same', input_shape=(28, 28, 1)),
        layers.BatchNormalization(),
        layers.Conv2D(32, kernel_size=(3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Dropout(0.25),
        
        # Second convolutional block
        layers.Conv2D(64, kernel_size=(3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.Conv2D(64, kernel_size=(3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Dropout(0.25),
        
        # Third convolutional block
        layers.Conv2D(128, kernel_size=(3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.Conv2D(128, kernel_size=(3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Dropout(0.25),
        
        # Flatten and dense layers
        layers.Flatten(),
        layers.Dense(256, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
        layers.BatchNormalization(),
        layers.Dropout(0.5),
        layers.Dense(10, activation='softmax')
    ])
    
    # Compile the model
    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=0.001),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    
    return model

def train_model(model, X_train, y_train, X_val, y_val, train_datagen, batch_size=128, epochs=30):
    """Train the neural network model with callbacks and data augmentation."""
    print("Starting model training...")
    start_time = time.time()
    
    # Create model checkpoint callback
    os.makedirs('models', exist_ok=True)
    checkpoint = ModelCheckpoint(
        'models/mnist_cnn_best.h5',
        monitor='val_accuracy',
        save_best_only=True,
        mode='max',
        verbose=1
    )
    
    # Create early stopping callback
    early_stopping = EarlyStopping(
        monitor='val_accuracy',
        patience=10,
        restore_best_weights=True,
        verbose=1
    )
    
    # Create learning rate reduction callback
    reduce_lr = ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.2,
        patience=3,
        min_lr=0.00001,
        verbose=1
    )
    
    # Callbacks list
    callbacks = [checkpoint, early_stopping, reduce_lr]
    
    # Train the model using data augmentation
    history = model.fit(
        train_datagen.flow(X_train, y_train, batch_size=batch_size),
        steps_per_epoch=len(X_train) // batch_size,
        epochs=epochs,
        validation_data=(X_val, y_val),
        callbacks=callbacks,
        verbose=1
    )
    
    training_time = time.time() - start_time
    print(f"Training completed in {training_time:.2f} seconds")
    
    return model, history

def evaluate_model(model, X_test, y_test):
    """Evaluate the trained neural network model."""
    print("Evaluating model performance...")
    
    # Evaluate on test set
    test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)
    print(f"Test accuracy: {test_accuracy:.4f}")
    print(f"Test loss: {test_loss:.4f}")
    
    # Make predictions
    y_pred_probs = model.predict(X_test)
    y_pred = np.argmax(y_pred_probs, axis=1)
    y_true = np.argmax(y_test, axis=1)
    
    # Generate classification report
    report = classification_report(y_true, y_pred)
    print("Classification Report:")
    print(report)
    
    # Compute confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    
    # Visualize confusion matrix
    plt.figure(figsize=(10, 8))
    plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title("Confusion Matrix")
    plt.colorbar()
    tick_marks = np.arange(10)
    plt.xticks(tick_marks, range(10))
    plt.yticks(tick_marks, range(10))
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    
    # Add text annotations to the confusion matrix
    thresh = cm.max() / 2
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            plt.text(j, i, format(cm[i, j], 'd'),
                     horizontalalignment="center",
                     color="white" if cm[i, j] > thresh else "black")
    
    plt.tight_layout()
    plt.show()
    
    # Plot training history
    if hasattr(model, 'history'):
        history = model.history.history
    else:
        return test_accuracy, report, cm
    
    plt.figure(figsize=(12, 5))
    
    # Plot accuracy
    plt.subplot(1, 2, 1)
    plt.plot(history['accuracy'], label='Train')
    plt.plot(history['val_accuracy'], label='Validation')
    plt.title('Model Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    
    # Plot loss
    plt.subplot(1, 2, 2)
    plt.plot(history['loss'], label='Train')
    plt.plot(history['val_loss'], label='Validation')
    plt.title('Model Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    
    plt.tight_layout()
    plt.show()
    
    return test_accuracy, report, cm

def save_model(model, filename='mnist_cnn_model'):
    """Save the trained model to disk."""
    # Save in Keras format
    model.save(f'{filename}.h5')
    
    # Convert to TFLite for mobile/embedded applications
    converter = tf.lite.TFLiteConverter.from_keras_model(model)
    tflite_model = converter.convert()
    with open(f'{filename}.tflite', 'wb') as f:
        f.write(tflite_model)
    
    print(f"Model saved as {filename}.h5 and {filename}.tflite")

def visualize_misclassifications(model, X_test, y_test, num_samples=15):
    """Visualize examples of misclassified digits."""
    # Get predictions
    y_pred_probs = model.predict(X_test)
    y_pred = np.argmax(y_pred_probs, axis=1)
    y_true = np.argmax(y_test, axis=1)
    
    # Find misclassified examples
    misclassified_indices = np.where(y_pred != y_true)[0]
    
    if len(misclassified_indices) == 0:
        print("No misclassifications found!")
        return
    
    # Select a subset of misclassified examples
    num_samples = min(num_samples, len(misclassified_indices))
    sample_indices = np.random.choice(misclassified_indices, num_samples, replace=False)
    
    # Create a figure to display them
    rows = int(np.ceil(num_samples / 5))
    plt.figure(figsize=(15, rows * 3))
    
    for i, idx in enumerate(sample_indices):
        plt.subplot(rows, 5, i + 1)
        plt.imshow(X_test[idx].reshape(28, 28), cmap='gray')
        plt.title(f"True: {y_true[idx]}, Pred: {y_pred[idx]}")
        plt.axis('off')
    
    plt.tight_layout()
    plt.show()

def main():
    """Main function to orchestrate the neural network model building process."""
    # Load and prepare data
    X_train, X_val, X_test, y_train, y_val, y_test = load_and_prepare_data()
    
    # Create data generators for augmentation
    train_datagen = create_data_generators(X_train)
    
    # Build the model
    model = build_cnn_model()
    print(model.summary())
    
    # Train the model
    model, history = train_model(model, X_train, y_train, X_val, y_val, train_datagen)
    
    # Evaluate model
    test_accuracy, report, cm = evaluate_model(model, X_test, y_test)
    
    # Visualize misclassifications
    visualize_misclassifications(model, X_test, y_test)
    
    # Save model
    save_model(model)
    
    print("Neural network model building completed successfully!")
    print(f"Final test accuracy: {test_accuracy:.4f}")
    
    return model

if __name__ == "__main__":
    main()
```
